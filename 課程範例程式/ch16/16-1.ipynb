```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import numpy.random as nr
```

```python
nr.seed(0)
plt.figure(dpi=100)
x1 = np.arange(-3, 7)
y1 = nr.normal(size=10)
x11 = np.arange(3, 13)
# x11 = nr.normal(size=10) * 10
y11 = nr.uniform(size=10) * 10
y2 = np.arange(10)
plt.plot(x1, y1, "go:", x11, y11, "bs-.", linewidth=2, markersize=4, label="s1")
plt.plot(y2, "k", marker="p", label="s2")
plt.xlim(-5, 15)
plt.xticks(range(-5, 16, 3))
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.annotate("A star", (7, 5))
plt.grid() #網格的意思
plt.legend()
plt.show()
```

```python
nr.seed(0)
plt.figure(dpi=100)
x1 = np.arange(-3, 7)
y1 = nr.normal(size=10)
x11 = nr.normal(size=10) * 10
y11 = nr.uniform(size=10) * 10
y2 = np.arange(10)
plt.plot(x1, y1, "go:", linewidth=2, markersize=4, label="s1")
plt.scatter(x11, y11, s=y11*15, c=y11, label="scatter1")
plt.plot(y2, "k*--", label="s2")
plt.xlim(-5, 15)
plt.xticks(range(-5, 16, 3))
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.annotate("A star", (7, 5))
plt.grid()
plt.legend()
plt.show()
```

```python
import yfinance as yf

tsmc = yf.download("6442.tw", start="2024-01-01", end="2024-12-31")
```

```python
tsmc["Change"] = tsmc.Close.diff()
tsmc.fillna(0, inplace=True)
```

```python
tsmc["Change_status"] = tsmc.Change.apply(lambda x: "r" if x > 0 else "g")
```

```python
plt.figure(dpi=150)
plt.scatter(tsmc.index, tsmc.Volume, s=np.abs(tsmc.Change)*10, c=tsmc.Change_status, alpha=0.6)
plt.xticks(rotation=30)
plt.show()
```

```python
np.corrcoef(tsmc.Change, tsmc.Volume)
```
